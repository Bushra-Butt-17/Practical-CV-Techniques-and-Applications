## <h2 style="color:green">Image 1: Black square with white inner square</h2>

This image consists of a black square with a white inner square.

```python
import numpy as np
import matplotlib.pyplot as plt

def create_image_with_square(image_size, box_size):
    """Creates an image with a black square and a white square inside it."""
    
    # Create a black image
    image = np.zeros((image_size, image_size, 3), dtype=np.uint8)
    
    # Calculate position for the inner white square
    start = (image_size - box_size) // 2
    end = start + box_size
    
    # Place the white square in the center
    image[start:end, start:end] = [255, 255, 255]  # White color

    # Show the image
    plt.imshow(image)
    plt.axis('off')  # Hide axis
    plt.show()

# Example usage
image_size = int(input("Enter image size: "))
box_size = int(input("Enter box size (size of inner square): "))
create_image_with_square(image_size, box_size)


```

## <h2 style="color:green">Image 2: Four small black squares at the corners</h2>

This image consists of four small black squares at the corners of a larger square.

```python
import numpy as np
import matplotlib.pyplot as plt

def create_image_with_four_boxes(image_size, box_size):
    """Creates an image with four black squares at the corners."""
    
    # Create a white image
    image = np.ones((image_size, image_size, 3), dtype=np.uint8) * 255
    
    # Ensure the box size is valid
    if box_size > image_size // 2:
        raise ValueError("Box size must be less than or equal to half of the image size.")

    # Place black boxes in each corner
    image[:box_size, :box_size] = [0, 0, 0]  # Top-left corner
    image[:box_size, -box_size:] = [0, 0, 0]  # Top-right corner
    image[-box_size:, :box_size] = [0, 0, 0]  # Bottom-left corner
    image[-box_size:, -box_size:] = [0, 0, 0]  # Bottom-right corner

    # Show the image
    plt.imshow(image)
    plt.axis('off')  # Hide axis
    plt.title("Image 2")
    plt.show()

# Example usage
image_size = int(input("Enter image size: "))  # E.g., 400
box_size = int(input("Enter box size: "))      # E.g., 50
create_image_with_four_boxes(image_size, box_size)


```


## <h2 style="color:green">Image 3: Grid of lines</h2>

This image consists of a grid of horizontal and vertical lines.

```python
def create_grid(image_size, num_lines, line_thickness):
    """Creates an image with a grid of lines."""
    
    # Create a white image
    image = np.ones((image_size, image_size, 3), dtype=np.uint8) * 255
    
    # Draw horizontal lines
    for i in range(num_lines):
        y = (i + 1) * (image_size // (num_lines + 1))
        image[y:y + line_thickness, :] = [0, 0, 0]  # Black color
    
    # Draw vertical lines
    for i in range(num_lines):
        x = (i + 1) * (image_size // (num_lines + 1))
        image[:, x:x + line_thickness] = [0, 0, 0]  # Black color

    # Show the image
    plt.imshow(image)
    plt.axis('off')  # Hide axis
    plt.show()

# Example usage
image_size = int(input("Enter image size: "))
num_lines = int(input("Enter number of lines: "))
line_thickness = int(input("Enter line thickness: "))
create_grid(image_size, num_lines, line_thickness)

```

<h1 style="color: #2E86C1;">Image Creation Task</h1>

<p style="color: #2C3E50;">This task involves creating an image with colored boxes in the corners. Here's a quick breakdown:</p>

<h2 style="color: #28B463;">1. White Image Creation</h2>
<p style="color: #2C3E50;">A white image is created using <code style="color: #C0392B;">numpy.ones()</code> and filled with <code style="color: #C0392B;">255</code> (white).</p>

<h2 style="color: #28B463;">2. Colored Boxes</h2>
<p style="color: #2C3E50;">Four small boxes (black, blue, green, red) are created using NumPy arrays and their respective RGB values are set:</p>
<ul>
    <li>Black: <code style="color: #C0392B;">[0, 0, 0]</code></li>
    <li>Blue: <code style="color: #C0392B;">[:, :, 2] = 255</code></li>
    <li>Green: <code style="color: #C0392B;">[:, :, 1] = 255</code></li>
    <li>Red: <code style="color: #C0392B;">[:, :, 0] = 255</code></li>
</ul>

<h2 style="color: #28B463;">3. Box Placement</h2>
<p style="color: #2C3E50;">Each box is placed in a corner of the image (top-left, top-right, bottom-left, bottom-right) by assigning slices of the image.</p>

<h2 style="color: #28B463;">4. Display and Save</h2>
<p style="color: #2C3E50;">The final image is displayed using <code style="color: #C0392B;">plt.imshow()</code> and saved to a file using <code style="color: #C0392B;">plt.imsave()</code>.</p>

</body>

```python
import numpy as np
import matplotlib.pyplot as plt

def create_image(size):
  """Creates a white image with colored boxes in the corners.

  Args:
    size: The size of the image (in pixels).

  Returns:
    A NumPy array representing the image.
  """

  # Create a white image
  image = np.ones((size, size, 3), dtype=np.uint8) * 255

  # Calculate the size of the colored boxes
  box_size = size // 10

  # Create the colored boxes
  black_box = np.zeros((box_size, box_size, 3), dtype=np.uint8)
  blue_box = np.zeros((box_size, box_size, 3), dtype=np.uint8)
  blue_box[:, :, 2] = 255
  green_box = np.zeros((box_size, box_size, 3), dtype=np.uint8)
  green_box[:, :, 1] = 255
  red_box = np.zeros((box_size, box_size, 3), dtype=np.uint8)
  red_box[:, :, 0] = 255

  # Place the colored boxes in the corners
  image[:box_size, :box_size] = black_box
  image[:box_size, -box_size:] = blue_box
  image[-box_size:, :box_size] = green_box
  image[-box_size:, -box_size:] = red_box

  return image

# Get the image size from the user
size = int(input("Enter the size of the image: "))

# Create the image
image = create_image(size)

# Display the image
plt.imshow(image)
plt.show()

# Save the image
plt.imsave("task2.png", image)
```

<h1 style="color: #2E86C1;">Image Mirroring Task</h1>

<p style="color: #212F3D;">This task involves reading an image, converting it to grayscale, and then mirroring its lower half to the upper half. Here's a step-by-step breakdown of the process:</p>

<h2 style="color: #28B463;">1. Reading the Image</h2>
<p style="color: #212F3D;">We use the <code style="color: #C0392B;">cv2.imread()</code> function from the OpenCV library to read the input image. If the image is loaded successfully, it proceeds to the next steps.</p>

<h2 style="color: #28B463;">2. Grayscale Conversion</h2>
<p style="color: #212F3D;">Once the image is read, it's converted into a grayscale image using <code style="color: #C0392B;">cv2.cvtColor()</code>. This simplifies the task since we only need to work with intensity values (one channel) instead of colors (three channels).</p>

<h2 style="color: #28B463;">3. Splitting the Image</h2>
<p style="color: #212F3D;">The height of the image is divided by 2, and the lower half of the image is extracted. This part will be used to mirror and replace the upper half.</p>

<h2 style="color: #28B463;">4. Mirroring the Lower Half</h2>
<p style="color: #212F3D;">The lower half of the grayscale image is flipped vertically using <code style="color: #C0392B;">cv2.flip()</code> to create a mirrored version. This mirrored half is then used to replace the upper half of the original image.</p>

<h2 style="color: #28B463;">5. Saving and Displaying the Image</h2>
<p style="color: #212F3D;">After mirroring, the modified image is saved using <code style="color: #C0392B;">cv2.imwrite()</code>. Additionally, the original and mirrored images are stacked side by side using <code style="color: #C0392B;">numpy.hstack()</code> and displayed in the Jupyter notebook using <code style="color: #C0392B;">matplotlib.pyplot</code>.</p>

```python
# Provide the input image path and output path
input_image_path = "C:\\Users\\PMLS\\Downloads\\puppy.jpg"  # Path to your input image
output_image_path = "mirrored_image.png"
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to mirror the lower half to the upper half of the image
def mirror_image(image_path, output_path):
    # Read the image
    image = cv2.imread(image_path)
    
    # Check if image is loaded correctly
    if image is None:
        print("Error: Unable to load image. Check the file path.")
        return
    
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Get the dimensions of the image
    height, width = gray_image.shape
    
    # If height is odd, reduce it by 1 to make it even
    if height % 2 != 0:
        height -= 1
        gray_image = gray_image[:height, :]
    
    # Get the lower half of the image
    lower_half = gray_image[height//2:, :]
    
    # Mirror the lower half (flip it vertically)
    mirrored_half = cv2.flip(lower_half, 0)
    
    # Replace the upper half of the image with the mirrored half
    mirrored_image = gray_image.copy()  # Create a copy to keep the original image
    mirrored_image[:height//2, :] = mirrored_half
    
    # Save the modified image to the disk
    cv2.imwrite(output_path, mirrored_image)
    
    # Stack the original and mirrored images horizontally
    side_by_side = np.hstack((gray_image, mirrored_image))
    
    # Display images using matplotlib in the notebook
    plt.figure(figsize=(10, 5))
    plt.imshow(side_by_side, cmap='gray')
    plt.axis('off')  # Hide axis
    plt.title('Original (Left) and Mirrored (Right) Image')
    plt.show()

# Call the function
mirror_image(input_image_path, output_image_path)

print(f"Mirrored image saved at: {output_image_path}")
```
